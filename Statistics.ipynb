```python
import pandas as pd
import numpy as np


df = pd.read_csv("drinks.csv")
df.head(10)
```

```python
# Beer_Servings

sum = 0
sd = 0
c = len(df["beer_servings"])
for i in df["beer_servings"]:
  sum = sum + i
mean_beer = sum/c
for j in df["beer_servings"]:
  a = (j - mean_beer)**2/c
  sd = a + sd
print(f"'beer_servings' column's mean is : {(mean_beer)}")
print(f"'beer_servings' column's standard deviation is : {(np.sqrt(sd))}")
print(f"'beer_servings' column's variance is : {(sd)}")
```

**2. From the given data, take a sample and find mean, standard deviation and variance for population in excel.
Also, validate Central Limit Theorem in excel and python as discussed in the class.**

```python
sample_1 = df[10:50]
sample_2 = df[50:90]
sample_3 = df[90:130]
sample_4 = df[70:110]
sample_5 = df[30:70]
```

```python
# Mean

a = np.mean(sample_1)
b = np.mean(sample_2)
c = np.mean(sample_3)
d = np.mean(sample_4)
e = np.mean(sample_5)

popmean = (a+b+c+d+e)/5
print(popmean)
```

```python
# Standard_Deviation

a = np.std(sample_1)
b = np.std(sample_2)
c = np.std(sample_3)
d = np.std(sample_4)
e = np.std(sample_5)


popstd = (a+b+c+d+e)/5
print(popstd)
```

```python
# Variance

a = np.var(sample_1)
b = np.var(sample_2)
c = np.var(sample_3)
d = np.var(sample_4)
e = np.var(sample_5)

popvar = (a+b+c+e+d)/5
print(popvar)
```
